```python
# Dolar-Track: Análisis de TRM semanal

def menu_dolar_track():
    trm_semanal = []
    while True:
        print("\n--- Menú Dolar-Track ---")
        print("1. Ingresar TRM de los últimos 7 días")
        print("2. Calcular Promedio y Volatilidad")
        print("3. Ver Reporte Ordenado")
        print("4. Salir")
        opcion = input("Seleccione una opción: ").strip()
        print(f"Opción recibida: '{opcion}'")  # Depuración

        if opcion == "1":
            trm_semanal = []
            for i in range(7):
                while True:
                    try:
                        valor = float(input(f"Ingrese la TRM del día {i+1}: "))
                        if valor < 0:
                            print("Error: La TRM no puede ser negativa.")
                        else:
                            trm_semanal.append(valor)
                            break
                    except ValueError:
                        print("Error: Ingrese un valor numérico válido.")
            print("TRM semanal registrada correctamente.")

        elif opcion == "2":
            if len(trm_semanal) != 7:
                print("Primero debe ingresar los 7 valores de TRM.")
                continue
            promedio = sum(trm_semanal) / 7
            dias_venta = 0
            dias_compra = 0
            for valor in trm_semanal:
                if valor > promedio:
                    dias_venta += 1
                elif valor < promedio:
                    dias_compra += 1
            volatilidad = max(trm_semanal) - min(trm_semanal)
            print(f"\nPromedio semanal: {promedio:.2f}")
            print(f"Días por encima del promedio (Alerta de Venta): {dias_venta}")
            print(f"Días por debajo del promedio (Alerta de Compra): {dias_compra}")
            print(f"Volatilidad (Máx - Mín): {volatilidad:.2f}")

        elif opcion == "3":
            if len(trm_semanal) != 7:
                print("Primero debe ingresar los 7 valores de TRM.")
                continue
            print("\nTRM de la semana (ordenada de mayor a menor):")
            for valor in sorted(trm_semanal, reverse=True):
                print(f"{valor:.2f}")

        elif opcion == "4":
            print("gracias por usar Dolar-Track.que tenga un buen día")
            break

        else:
            print("Opción no válida. Intente de nuevo.")

menu_dolar_track()
```
