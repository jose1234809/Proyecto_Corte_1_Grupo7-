```python
# ğŸ® SERVICIO: Sistema de GestiÃ³n de Precios - Tienda de Videojuegos "GAME ZONE"
# ============================================================================
# CONTEXTO: "GAME ZONE" es una tienda especializada en videojuegos que vende
# consolas de actualidad (PlayStation 5, Xbox Series X, Nintendo Switch Pro),
# videojuegos AAA (The Legend of Zelda, Final Fantasy XVI, Starfield) y
# controles compatibles. Los precios llegan desordenados del distribuidor y
# necesitan organizarse para mostrar en catÃ¡logo online.
#
# FIN ESPECÃFICO: Organizar el inventario por rangos de precio y calcular
# el costo final con IVA (19%) para que los clientes vean el precio exacto
# que pagarÃ¡n en checkout. Esto mejora la experiencia de compra y la
# transparencia de la tienda.
# ============================================================================

print("\n" + "ğŸ® " * 20)
print("ğŸ® BIENVENIDO AL SISTEMA DE GESTIÃ“N DE PRECIOS - GAME ZONE ğŸ®")
print("ğŸ® " * 20)

print("\nğŸ“– CONTEXTO DEL SERVICIO:")
print("-" * 70)
print("Usted estÃ¡ utilizando el Sistema de GestiÃ³n de Precios de GAME ZONE")
print("una tienda especializada en videojuegos y accesorios.")
print("\nğŸ¯ OBJETIVO:")
print("   â€¢ Organizar productos por rango de precio (menor a mayor)")
print("   â€¢ Calcular el precio final con IVA (19%) para cliente")
print("   â€¢ Transparencia total en el proceso de venta")
print("\nğŸ“¦ PRODUCTOS EN INVENTARIO:")
print("   Â¡No lo dudes, tenemos la consola de tus sueÃ±os y muchos mÃ¡s!")
print("\n   â€¢ Consolas de Ãºltima generaciÃ³n: PlayStation 5, Xbox Series X, Nintendo Switch")
print("   â€¢ Videojuegos AAA: The Legend of Zelda, Final Fantasy XVI, Starfield")
print("   â€¢ Controles compatibles para todas las consolas")
print("\nğŸ’¼ PROCESO:")
print("   1. Estaremos recibiendo productos desordenados del proveedor")
print("   2. Por consiguiente, los ordenaremos por precio de menor a mayor para mostrar un catÃ¡logo mÃ¡s ordenado")
print("   3. Posteriormente, Â¡Aplicaremos el IVA del 19% a cada producto para enseÃ±arte nuestros precios de locura!")
print("   4. Finalmente, mostraremos un catÃ¡logo final con los precios definitivos de venta")
print("-" * 70)

print("\nğŸ® INICIANDO SERVICIO DE GESTIÃ“N DE PRECIOS - GAME ZONE")
print("=" * 70)

# 1. PASO 1: Lista desordenada de precios de productos de videojuegos
# Estos precios incluyen: consolas, videojuegos premium y controles
productos = [
    ("Nintendo Switch Pro", 350),
    ("Control Xbox Series X", 70),
    ("The Legend of Zelda: Tears of the Kingdom", 80),
    ("PlayStation 5 Standard", 550),
    ("Control DualSense PS5", 85),
    ("Starfield (Xbox)", 75),
    ("Control Nintendo Pro", 65),
    ("Xbox Series X", 500),
    ("Final Fantasy XVI", 90),
    ("Control Extra Xbox", 68)
]

# Separar productos y precios
nombres_productos = [p[0] for p in productos]
precios_originales = [p[1] for p in productos]

print("\nğŸ“¥ Productos recibidos del distribuidor (SIN ORDENAR por precio):")
for i, (nombre, precio) in enumerate(productos, 1):
    print(f"   {i}. {nombre}: ${precio}")

# 2. PASO 2: Ordenar los precios de menor a mayor
# CrÃ©amos tuplas para mantener la relaciÃ³n producto-precio
productos_con_precios = list(zip(nombres_productos, precios_originales))
productos_con_precios.sort(key=lambda x: x[1])

nombres_ordenados = [p[0] for p in productos_con_precios]
precios_ordenados = [p[1] for p in productos_con_precios]

print("\nâœ… Productos ORDENADOS por rango de precio (menor a mayor):")
for i, (nombre, precio) in enumerate(productos_con_precios, 1):
    print(f"   {i}. {nombre}: ${precio}")

# 3. PASO 3: Aplicar IVA (19%) a cada precio ordenado
# El cliente verÃ¡ este precio final en el carrito
print("\nğŸ’° Calculando precios finales con IVA (19%) para mostrar a clientes...")
precios_con_iva = []
for precio in precios_ordenados:
    precio_con_iva = precio * 1.19  # Aplicar 19% de IVA
    precios_con_iva.append(precio_con_iva)

# 4. PASO 4: Mostrar los resultados finales del servicio
print("\n" + "=" * 70)
print("ğŸ“‹ CATÃLOGO FINAL - PRECIOS CON IVA INCLUIDO")
print("=" * 70)

print("\nğŸ”· Lista de precios base (sin IVA) - ORDENADA:")
print(f"   {precios_ordenados}")

print("\nğŸ’³ Productos con precio FINAL DE VENTA (IVA 19% incluido):")
print("-" * 70)

precios_formateados = [f"{precio:.2f}" for precio in precios_con_iva]
for i, (nombre, precio_base, precio_iva) in enumerate(zip(nombres_ordenados, precios_ordenados, precios_formateados), 1):
    categoria = "ğŸ•¹ï¸ CONSOLA" if "PlayStation" in nombre or "Xbox" in nombre or "Switch" in nombre else "ğŸ® JUEGO" if "Zelda" in nombre or "Final Fantasy" in nombre or "Starfield" in nombre else "ğŸ¯ CONTROL"
    print(f"   {i}. {categoria} | {nombre}")
    print(f"      Base: ${precio_base:.2f} â†’ Precio Final: ${precio_iva}")
    print()

print("=" * 70)
print("âœ¨ Servicio completado - CatÃ¡logo actualizado y disponible para venta")
print("=" * 70)

# 5. PASO 5: INTERACCIÃ“N CON EL CLIENTE - PROCESO DE COMPRA
print("\n" + "ğŸ›’ " * 20)
print("ğŸ›’ BIENVENIDO AL PROCESO DE COMPRA - GAME ZONE ğŸ›’")
print("ğŸ›’ " * 20)

# Carrito de compras
carrito = []
total_carrito = 0

# Proceso de compra repetitivo
comprar_mas = True

while comprar_mas:
    print("\nğŸ“‹ SELECCIONA UN PRODUCTO PARA COMPRAR:")
    print("-" * 70)
    for i, (nombre, precio) in enumerate(productos, 1):
        precio_con_iva_producto = precio * 1.19
        categoria = "ğŸ•¹ï¸ CONSOLA" if "PlayStation" in nombre or "Xbox" in nombre or "Switch" in nombre else "ğŸ® JUEGO" if "Zelda" in nombre or "Final Fantasy" in nombre or "Starfield" in nombre else "ğŸ¯ CONTROL"
        print(f"   [{i}] {categoria} | {nombre} â†’ ${precio_con_iva_producto:.2f}")

    print("\n   [0] Finalizar compra y ver resumen")
    print("-" * 70)

    # Obtener selecciÃ³n del usuario
    while True:
        try:
            seleccion = int(input("\nğŸ‘‰ Ingresa el nÃºmero del producto que deseas comprar (o 0 para finalizar): "))
            
            if seleccion == 0:
                comprar_mas = False
                break
            
            if 1 <= seleccion <= len(productos):
                producto_seleccionado = productos[seleccion - 1][0]
                precio_base_seleccionado = productos[seleccion - 1][1]
                precio_final_seleccionado = f"{precio_base_seleccionado * 1.19:.2f}"
                iva_calculado = float(precio_final_seleccionado) - precio_base_seleccionado
                
                categoria = "ğŸ•¹ï¸ CONSOLA" if "PlayStation" in producto_seleccionado or "Xbox" in producto_seleccionado or "Switch" in producto_seleccionado else "ğŸ® JUEGO" if "Zelda" in producto_seleccionado or "Final Fantasy" in producto_seleccionado or "Starfield" in producto_seleccionado else "ğŸ¯ CONTROL"
                
                print("\n" + "=" * 70)
                print("âœ… PRODUCTO SELECCIONADO")
                print("=" * 70)
                print(f"\n{categoria} Producto: {producto_seleccionado}")
                print(f"ğŸ’µ Precio base: ${precio_base_seleccionado:.2f}")
                print(f"ğŸ“Š IVA (19%): ${iva_calculado:.2f}")
                print(f"ğŸ’³ Precio final (con IVA): ${precio_final_seleccionado}")
                
                # ConfirmaciÃ³n de compra
                confirmacion = input("\nÂ¿Deseas proceder con la compra? (si/no): ").lower()
                
                if confirmacion in ['si', 's', 'yes', 'y']:
                    # Preguntar cantidad
                    while True:
                        try:
                            cantidad = int(input("\nÂ¿CuÃ¡ntas unidades deseas comprar? (1-10): "))
                            if 1 <= cantidad <= 10:
                                break
                            else:
                                print("âŒ Por favor, ingresa una cantidad entre 1 y 10.")
                        except ValueError:
                            print("âŒ Error: Debes ingresar un nÃºmero vÃ¡lido.")
                    
                    # Agregar al carrito
                    subtotal = float(precio_final_seleccionado) * cantidad
                    carrito.append({
                        'categoria': categoria,
                        'nombre': producto_seleccionado,
                        'precio_unitario': float(precio_final_seleccionado),
                        'cantidad': cantidad,
                        'subtotal': subtotal
                    })
                    total_carrito += subtotal
                    
                    print("\n" + "âœ¨ " * 20)
                    print(f"âœ¨ Â¡{cantidad} unidad(es) agregada(s) al carrito!")
                    print("âœ¨ " * 20)
                    print(f"\n   Total en carrito: ${total_carrito:.2f}")
                    
                    # Preguntar si comprar mÃ¡s
                    comprar_otro = input("\nÂ¿Deseas comprar otro producto? (si/no): ").lower()
                    if comprar_otro not in ['si', 's', 'yes', 'y']:
                        comprar_mas = False
                    
                    break
                else:
                    print("\nâŒ Compra cancelada. Vuelve a intentarlo si lo deseas.\n")
                    break
            else:
                print("\nâŒ OpciÃ³n invÃ¡lida. Por favor, selecciona una opciÃ³n vÃ¡lida.")
        
        except ValueError:
            print("\nâŒ Error: Debes ingresar un nÃºmero vÃ¡lido.")
        except KeyboardInterrupt:
            print("\n\nğŸ‘‹ Â¡Hasta pronto! Gracias por visitarnos en GAME ZONE.")
            comprar_mas = False
            break

# Mostrar resumen final si hay items en el carrito
if carrito:
    print("\n" + "ğŸ‰ " * 20)
    print("ğŸ‰ Â¡RESUMEN FINAL DE COMPRAS - GAME ZONE! ğŸ‰")
    print("ğŸ‰ " * 20)
    
    print("\nğŸ“¦ DETALLES DE TU CARRITO:")
    print("=" * 70)
    
    for idx, item in enumerate(carrito, 1):
        print(f"\n{idx}. {item['categoria']} {item['nombre']}")
        print(f"   Precio unitario: ${item['precio_unitario']:.2f}")
        print(f"   Cantidad: {item['cantidad']} unidad(es)")
        print(f"   Subtotal: ${item['subtotal']:.2f}")
    
    print("\n" + "=" * 70)
    print(f"ğŸ’³ TOTAL A PAGAR: ${total_carrito:.2f}")
    print("=" * 70)
    
    print("\nâœ¨ Tu compra ha sido procesada correctamente.")
    print("ğŸ“§ Te enviaremos un correo de confirmaciÃ³n en breve.")
    print("ğŸšš Tus productos serÃ¡n enviados en 2-3 dÃ­as hÃ¡biles.")
    print("\nÂ¡Gracias por tu compra en GAME ZONE!")
else:
    print("\n" + "ğŸ›’ " * 20)
    print("No realizaste ninguna compra.")
    print("ğŸ‘‹ Gracias por visitarnos en GAME ZONE. Â¡Vuelve pronto!")
    print("ğŸ›’ " * 20)
```
